```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine_data=pd.read_csv('wine.csv')
```

```python
print(wine_data.describe())
wine_data.head()
```

```python
wine_data
```

```python
wine_data.shape
```

```python
wine_data.info()
```

```python
wine_data_ary=wine_data.values
wine_data_ary
```

```python
wine_data_norm=scale(wine_data_ary)
wine_data_norm
```

```python
pca = PCA()
pca_data_values = pca.fit_transform(wine_data_norm)
pca_data_values
```

```python
pca.components_
```

```python
variance = pca.explained_variance_ratio_
variance
```

```python
Variance = np.cumsum(np.round(variance,decimals= 4)*100)
Variance
```

```python
plt.plot(Variance,color="red");
```

```python
final_df=pd.concat([wine_data['Type'],pd.DataFrame(pca_data_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
import seaborn as sns
fig=plt.figure(figsize=(18,15))
sns.scatterplot(data=final_df);
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type');
```

Cheacking with other Clustering Algorithm

1) ......Hierarchical Clustring......

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_data_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_data_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine_data2=wine_data.copy()
wine_data2['clustersid']=hclusters.labels_
wine_data2
```

2) ........K Means Clustering.........

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_data_norm)
    wcss.append(kmeans.inertia_)
    import warnings
    warnings.filterwarnings('ignore')
```

```python
plt.plot(range(3,8),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

# k=3

```python
clusters=KMeans(3,random_state=30).fit(wine_data_norm)
clusters
```

```python
clusters.labels_
```

```python
wine_data3=wine_data.copy()
wine_data3['clustersid']=clusters.labels_
wine_data3
```

```python
wine_data3['clustersid'].value_counts()
```
